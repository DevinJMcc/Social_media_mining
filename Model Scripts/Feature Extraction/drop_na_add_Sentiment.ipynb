```python
df = pd.read_csv('project13-sideeffects - NEW.csv', engine='python')
df = df.drop(axis=0, columns=['H/NH',	'P/PSA',	'B/A',	'N/S/J/G', 'Unnamed: 10', 'name','profile_url','description'])
df = df.dropna(axis=0,subset=['V/L'])
df = df.dropna()
df
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

def sentiment_detect(sentence):
 
    sta = SentimentIntensityAnalyzer()
 
    sta_dict = sta.polarity_scores(sentence)
 
    # decide sentiment as positive, negative and neutral
    if sta_dict['compound'] >= 0.05 :
        return 'Positive'
 
    elif sta_dict['compound'] <= - 0.05 :
        return "Negative"
 
    else :
        return "Neutral"
```

```python
sentiment = []
for item in df.text.to_list():
  sentiment += [sentiment_detect(item)]

df['sentiment'] = sentiment
df
```
